```python
fuzzy_ans = ['' for i in range(len(ques))]
op = 0
for q1 in keys_1:
    max_score = 0
    for q2 in keys_2:
        score = fuzz.ratio(q1,q2)
        if(score>=90):
            q = q2
            ans = dic_1[q1]
            break
        elif(score>max_score):
            max_score = score
            ans = dic_1[q1]
            q = q2
    ind = dic_2[q]
    for i in ind:
        fuzzy_ans[i] = ans
    if(op%500 == 0):
        print('op = ',op)
    op+=1
```
